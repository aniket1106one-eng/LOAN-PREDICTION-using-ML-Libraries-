```python
from sklearn.preprocessing import MinMaxScaler
x=MinMaxScaler().fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test , y_train, y_test = train_test_split ( x , y ,test_size=0.3 , random_state=45)
print(x.shape)
print(y.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
### APPLY LOGESTIC REGRESSION MODEL 
```

```python
from sklearn.linear_model import LogisticRegression
lg= LogisticRegression(solver='saga' ,random_state=1 , max_iter=500)
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix ,classification_report
from sklearn.metrics import accuracy_score
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
Lracc=(accuracy_score(y_pred,y_test))
Lracc
print('LOGESTIC REGRESSION ACCURACY:{:.2f}%'.format(Lracc*100))
      
      
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred= knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test)*100)
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
KnnScore=[]
for i in range(1,15):
    Knn = KNeighborsClassifier(n_neighbors=i)
    Knn.fit(x_train,y_train)
    KnnScore.append(Knn.score(x_test,y_test))

plt.plot(range(1,15),KnnScore)
plt.xticks(np.arange(1,15,1))


plt.show()
```

```python
knn= max(KnnScore)
print("KNN ACCURACY : {:.2f}%".format(knn*100))
```

```python
#### SUPPORT VECTOR MACHINE
```

```python
from sklearn.svm import SVC
svc= SVC(kernel='rbf', max_iter=600)
svc.fit(x_train, y_train)
```

```python
y_pred =svc.predict(x_test)
y_pred
```

```python
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
sv=(accuracy_score(y_pred,y_test))
print('SVC ACCURACY : {:.2f}%'.format(sv*100))
```

```python
#### DECSION TREE 
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier().fit(x_train,y_train)
dtc.fit(x_train,y_train)

```

```python
y_pred=dtc.predict(x_test)
y_pred
```

```python
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
dtc=accuracy_score(y_pred,y_test)
print('DECISION TREE :{:.2f}%'.format(dtc))
```

```python
dtcScore=[]
for i in range(2,30):
    dtc= DecisionTreeClassifier(max_leaf_nodes=i)
    dtc.fit(x_train,y_train)
    dtcScore.append(dtc.score(x_test,y_test))

plt.plot(range(2,30),dtcScore)
plt.xticks(np.arange(2,30,1))


plt.show()
```

```python
dtcacc= max(dtcScore)
print('MAXIMUM DECISION TREE ACC :{:.2f}%'.format(dtcacc*100))
```

```python
#### RANDOM FOREST
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
```

```python
y_pred= rfc.predict(x_test)
y_pred
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
rnracc=[]
for i in range(2,30):
    rfc= RandomForestClassifier(n_estimators=100,max_leaf_nodes=i).fit(x_train,y_train)
    rnracc.append(rfc.score(x_test,y_test))

plt.plot(range(2,30),rnracc)
plt.xticks(np.arange(2,30,1))
plt.grid(True)

plt.show()

```

```python
rnc = max(rnracc) * 100
print('RANDOM FOREST ACCURACY: {:.2f}%'.format(rnc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc= GradientBoostingClassifier(subsample=0.5,
                                n_estimators=500,
                                max_leaf_nodes=12,
                                max_depth=4)
gbc.fit(x_train,y_train)
```

```python
y_pred= gbc.predict(x_test)
y_pred
```

```python
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
gb=(accuracy_score(y_pred, y_test)*100)
print('GRADIENT BOOSTING ACCURACY ;{:.2f}%'.format(gb))
```
